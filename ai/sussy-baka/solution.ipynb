```python
import pandas as pd
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
X = df[['x', 'y']]
```

```python
dbscan = DBSCAN()
```

```python
dbscan.fit(X)
```

```python
#plot the data
plt.scatter(X['x'], X['y'], c=dbscan.labels_)
```

```python
#print the  name of outliers
print(df['point'][dbscan.labels_ == -1])
```

```python
#print in format igeneer{outlier1_outlier2_...}
print('Ingeneer{' + '_'.join(df['point'][dbscan.labels_ == -1]) + '}')
```
